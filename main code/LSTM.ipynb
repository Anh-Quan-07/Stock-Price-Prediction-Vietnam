## 3. Preprocess data

```python
from matplotlib.dates import YearLocator, DateFormatter, MonthLocator # Thêm MonthLocator vào để sửa lỗi

# Chuyển đổi cột "Ngày" sang dạng datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Sắp xếp lại dữ liệu theo thứ tự thời gian
df = df.sort_values(by='Date')

# Chuyển đổi định dạng các cột giá thành số thực
df['Closing Price'] = df['Closing Price'].replace(',', '').astype(float)
df['Opening Price'] = df['Opening Price'].replace(',', '').astype(float)
df['Highest Price'] = df['Highest Price'].replace(',', '').astype(float)
df['Lowest Price']  = df['Lowest Price'].replace(',', '').astype(float)

# Lấy thông tin năm từ cột "Ngày"
df['Năm'] = df['Date'].dt.year # chuyển đổi cột ngày sang năm

# Tạo đồ thị giá đóng cửa qua các năm
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Closing Price'], label='Closing Price', color='red')
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('HCM closing price chart over the years')
plt.legend(loc='best')

# Định dạng đồ thị hiển thị các ngày tháng theo năm-tháng
years = YearLocator()
yearsFmt = DateFormatter('%Y')
months = MonthLocator()  # Thêm dòng này để khai báo MonthLocator
plt.gca().xaxis.set_major_locator(years)
plt.gca().xaxis.set_major_formatter(yearsFmt)
plt.gca().xaxis.set_minor_locator(months)

plt.tight_layout() # loại bỏ những thành phần trùng lập
plt.show()
```

```python
# lấy cột đóng cửa để xử lý
df1 = pd.DataFrame(df,columns=['Date','Closing Price']) # tạo 1 dataframe df1 gồm 2 cột từ df
df1.index = df1.Date                                    # Thiết lập cột Date làm index cho DataFrame df1. Điều này giúp mỗi hàng có một chỉ mục là giá trị ngày tháng tương ứng.
df1.drop('Date',axis=1,inplace=True)                    # xóa cột Date khỏi các cột của DataFrame, vì giá trị này đã được chuyển sang làm index.Tham số axis=1 chỉ định rằng bạn muốn xóa một cột, và inplace=Truethay đổi trực tiếp trên DataFrame df1.
df1
```

```python
# Tính toán chỉ số MA (Moving Average)
df1['MA50'] = df1['Closing Price'].rolling(window=50).mean()
df1['MA200'] = df1['Closing Price'].rolling(window=200).mean()
```

```python
# Tính toán chỉ số RSI (Relative Strength Index)
def compute_rsi(data, window=14):
    delta = data.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=window, min_periods=1).mean()
    avg_loss = loss.rolling(window=window, min_periods=1).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))

    return rsi

df1['RSI'] = compute_rsi(df1['Closing Price'])
```

```python
# Biểu đồ RSI
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 2)
plt.plot(df1['RSI'], label='RSI', color='purple', alpha=0.7)
plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
plt.title('Relative Strength Index (RSI)')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#chia tập dữ liệu
df11 = pd.DataFrame(df,columns=['Date','Closing Price'])
df11.index = df11.Date                                    # Thiết lập cột Date làm index cho DataFrame df1. Điều này giúp mỗi hàng có một chỉ mục là giá trị ngày tháng tương ứng.
df11.drop('Date',axis=1,inplace=True)                    # xóa cột Date khỏi các cột của DataFrame, vì giá trị này đã được chuyển sang làm index.Tham số axis=1 chỉ định rằng bạn muốn xóa một cột, và inplace=Truethay đổi trực tiếp trên DataFrame df1.

data = df11.values   #Trả về toàn bộ giá trị của DataFrame df1 dưới dạng một mảng NumPy (numpy array),
train_data = data[:1252]
test_data = data[1252:]
data
```

```python
#chuẩn hóa dữ liệu(Tăng tốc độ huấn luyện,Giảm ảnh hưởng của các giá trị cực đại,Cải thiện độ chính xác,hồi quy nhanh hơn,.....)
sc = MinMaxScaler(feature_range=(0,1))
sc_train = sc.fit_transform(data)
sc_train
```

```python
#tạo vòng lặp các giá trị
x_train,y_train=[],[]
for i in range(50,len(train_data)):
  x_train.append(sc_train[i-50:i,0]) #lấy 50 giá đóng cửa liên tục
  y_train.append(sc_train[i,0]) #lấy ra giá đóng cửa ngày hôm sau
```

```python
#xếp dữ liệu thành 1 mảng 2 chiều
x_train = np.array(x_train)
y_train = np.array(y_train)

#xếp lại dữ liệu thành mảng 1 chiều
x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
y_train = np.reshape(y_train,(y_train.shape[0],1))
```

## 4. Build model

```python
#xây dựng mô hình
model = Sequential() #tạo lớp mạng cho dữ liệu đầu vào
#2 lớp LSTM
model.add(LSTM(units=128,input_shape=(x_train.shape[1],1),return_sequences=True))
model.add(LSTM(units=64))                                                           # gồm 2 lớp
model.add(Dropout(0.5)) #loại bỏ 1 số đơn vị tránh học tủ (overfitting)
model.add(Dense(1)) #output đầu ra 1 chiều
#đo sai số tuyệt đối trung bình có sử dụng trình tối ưu hóa adam
model.compile(loss='mean_absolute_error',optimizer='adam')
```

## 5. Train model

```python
#huấn luyện mô hình
save_model = "/content/gdrive/MyDrive/Khóa luân 2024 (Nguyễn Tiến Anh QUân)/allsavemodel/savemodelLSTM/save_model.keras"
best_model = ModelCheckpoint(save_model,monitor='loss',verbose=2,save_best_only=True,mode='auto')
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=2)
#model.fit(x_train,y_train,epochs=50,batch_size=50,verbose=2,callbacks=[best_model])

# K-fold Cross Validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

for train_index, val_index in kfold.split(x_train):
    x_train_fold, x_val_fold = x_train[train_index], x_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # Huấn luyện mô hình với EarlyStopping và ModelCheckpoint
    model.fit(x_train_fold, y_train_fold, epochs=50, batch_size=50, verbose=2,
              validation_data=(x_val_fold, y_val_fold), callbacks=[best_model, early_stopping])
```

```python
#dữ liệu train
y_train = sc.inverse_transform(y_train) #giá thực(từ chuẩn hóa về dạng thực tế)
final_model = load_model("/content/gdrive/MyDrive/Khóa luân 2024 (Nguyễn Tiến Anh QUân)/allsavemodel/savemodelLSTM/save_model.keras")
y_train_predict = final_model.predict(x_train) #dự đoán giá đóng cửa trên tập đã train
y_train_predict = sc.inverse_transform(y_train_predict) #chuyển đổi giá dự đoán về lại giá thực
```
